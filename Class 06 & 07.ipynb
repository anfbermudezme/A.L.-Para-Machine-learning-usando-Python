```python
A = np.array([[3, 7], [5, 2]])
print(A)
```

```python
print('Circulo unitario')
graficarmatriz(np.array([[1, 0], [0, 1]]))

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()

print('Circulo unitario transformado')
graficarmatriz(A)

plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
U, D, V = np.linalg.svd(A)
print(U)
```

```python
print('Circulo unitario')
graficarmatriz(np.array([[1, 0], [0, 1]]))

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()

# La matriz V es la primer transformacin y es una rotacin

print('Primer rotacin V')
graficarmatriz(np.array(V))

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()

# Vemos lo que hace la matriz diagonal D

print('Escala (D)')
graficarmatriz(np.diag(D).dot(V))

plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()

# La segunda rotacin viene dada por U

print('Segunda rotacin U')
graficarmatriz(U.dot(np.diag(D).dot(V)))

plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()
```
